### Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

## Modelling - What is our baseline?
What is the simplest model we can think of?

In this case, if our model is randomly guessing the species of flower, or guessing a single species for every data points, we would expect to have the model accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced.

So, our model should atleast beat 33% accuracy

### Simple manual model

```python
def single_feature_prediction(petal_length):
    ''' Predicts the Iris species given the petal length'''
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
X_train[:,2]
```

```python
manual_y_prediction = np.array([single_feature_prediction(val) for val in X_train[:,2]])
```

```python
np.mean(manual_y_prediction == y_train)
```

### Modelling - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=200)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train) # We do not calculate score on the same data used during training
```

```python
model.score(X_test, y_test)
```

#### Using validation set to evaluate our model

```python
Xt, Xv, yt, yv = train_test_split(X_train, y_train, test_size = 0.25, random_state = 0)
```

```python
modelf = LogisticRegression()
```

```python
modelf.fit(Xt, yt)
```

```python
y_pred = modelf.predict(Xv)
```

```python
np.mean(y_pred == yv)
```

```python
modelf.score(Xv, yv)
```

#### Using cross-validation to evaluate our model

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
```

```python
logreg = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(estimator = logreg, X = X_train, y = y_train, cv =5, scoring = 'accuracy')
```

```python
np.mean(accuracies)
```

### Hyperparameter Tuning

```python
for regpram in (8, 9, 9.1, 9.2, 9.4, 9.7, 9.9, 10):
    print(regpram)
    model = LogisticRegression(max_iter=200, C=regpram)
    accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring = "accuracy")
    print(np.mean(accuracies))
```

### Final Model

```python
model = LogisticRegression(max_iter=200, C=9)
```

```python
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test)
```

```python
# This is Final test accuracy
print(np.mean(y_test_pred == y_test)) # Test set accuracy
```

```python
y_test_pred
```

```python
labelencoder_y.inverse_transform(y_test_pred)
```
